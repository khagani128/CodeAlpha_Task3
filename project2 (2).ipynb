```python
plt.figure(figsize=(10,6))
colors=['#4ECDC4' if gender=='Male' else '#FF6B6B' for gender in df['Gender']]
bars=plt.bar(df['Gender'],df['Exam_Score'],color=colors)
plt.title('Sagirdlerin imtahan bali')
plt.ylabel('imtahan bali',color='#4ECDC4')
plt.xticks(rotation=45)
plt.grid(True,alpha=0.3)
plt.show()
```

```python
                
colors=[]
for score in df['Exam_Score']:
    if score<50:
        colors.append('#4ECDC4')
    elif score>50 and score<71:
        colors.append('#FF6B6B')
    elif score>70 and score<91:
        colors.append('#DDA0DD')
    elif score>90:
        colors.append('#96CEB4')
        
plt.figure(figsize=(10,6))
bars=plt.bar(df['Exam_Score'],df['Family_Income'],color=colors)
plt.title('Sagirdlerin imtahan balina gore siralanmasi')
plt.xlabel('Imtahan bali')
plt.xticks(rotation=45)

plt.show()
```
